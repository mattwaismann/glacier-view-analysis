```python
import seaborn as sns
import numpy as np
import statsmodels.api as sm
import math
import datetime

def plot_areas(rel_error = 8, log=True, all_areas=False, weighted=False):
    areas = ['South America', 'Asia', 'Europe', 'North America', 'Caucausus',
           'Oceania']
    better_glaciers = training_err[training_err["rel_error_train"]<rel_error]
    # better_glaciers = predicted_glaciers
    # predicted_no_thresh.index=predicted_no_thresh["Dates"]
    for area in areas[:-1]:
        # try:
            plt.figure(figsize=(10,5))
            selected_glaciers = geo_areas.groupby("geog_area_rollup").get_group(area)["glac_id"].to_list()

            selected_glaciers = list(set(selected_glaciers).intersection(set(predicted_glaciers)))

            final_filtered_glaciers = []
            for i in selected_glaciers:
                # if i in list(better_glaciers["glims_id"]):
                if i in list(predicted_no_thresh.columns):
                    final_filtered_glaciers.append(i)
            final_filtered_glaciers
            selected_glaciers = final_filtered_glaciers
            num_glaciers = len(selected_glaciers)
            
            
            
            
            # if all_areas:
            cols = len(predicted_no_thresh[selected_glaciers].columns)
            f = predicted_no_thresh[selected_glaciers].mean(axis=1)
            f = f.dropna()
            
            if area=="South America":
                f = f[f.index.year>2000]
                
            
            
            # print(cols)
            # print(predicted_no_thresh[selected_glaciers].isna().sum(axis=1))

            if area=="South America":
                weights = (cols -  predicted_no_thresh[selected_glaciers][predicted_no_thresh.index.year>2000].isna().sum(axis=1))/cols
            else:
                weights = (cols -  predicted_no_thresh[selected_glaciers].isna().sum(axis=1))/cols
            weights = weights[weights!=0.0]
            weights = weights.to_list()[:len(f)]
            # print(weights)


            X = (f.index.astype(int)).to_list()
            y = f.to_list()
            
            

            new_y = []

            for i in range(len(y)):
                if y[i]==0.0:
                    weights.pop(i)
                    X.pop(i)
                else:
                    new_y.append(y[i])
            y = new_y
            Y = y.copy()
            

            log_y = [math.log(i) for i in y]

            # Calculate weights based on the inverse of the variance of the errors
            weights

            # Fit weighted least squares regression model
            X = np.array(X)
            x = sm.add_constant(X)

            model = sm.WLS(log_y, x, weights=weights)
            results = model.fit()
            
            
            constant, slope = results.params

            # Given slope (m) and intercept (b)
            m = slope # Example slope
            b = constant  # Example intercept

            # Generate x values (non-logarithmic scale)
            # x = np.linspace(1984, 2024, 10)
            x = np.array(x)

            # Calculate y values using the equation y = e^(m * x + b)
            
            
            # Plot the non-logarithmic line
            # plt.figure(figsize=(8, 6))
            plt.rcParams["figure.figsize"] = (10,8)
            
            if not log:
                y = np.exp(m * x + b)
                plt.ylabel("Area (Non-Log Scale)")
                plt.plot(X,Y,alpha=0.3)
                print("\n\nPercentage decline since beginning: ",round((100*(y[-1]-y[0])/y[0])[-1],4),"% \n\n")
                
            else:
                y = (m * x + b)
                plt.ylabel("Area (Log Scale)")
                plt.plot(X,log_y,alpha=0.3)
                
            plt.plot(x, y,lw=1,color="blue")
            plt.plot([x[0],x[-1]],[y[0],y[-1]], alpha=0.2)
            

            # Adding labels and title
            plt.xlabel("Years")
            # plt.ylabel("Area (Non-Log Scale)")
            plt.title(f"Average area (km2) of {num_glaciers} glaciers in {area} ")
            # plt.legend()

            # Show the plot
            plt.grid(True)
            plt.xticks([i for i in range(0,60,5)],[i+1970 for i in range(0,60,5)])
            plt.show()
            # print(results.summary())
        # except:
        #     pass
            
                

                
plot_areas(rel_error = 200, log=False)
pass

# Superimpose individual curves - with limited opacity, and average on top with thickness
# Keep x axis range fixed for all plots - DONE

# For non-NaN values plot confidence interval of the mean - 2SD range - DONE
# Debiase the model predicitons by aligning predicted area with the ground truth at training point - NA
```
